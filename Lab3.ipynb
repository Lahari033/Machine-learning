```python
m=np.random.randn()
c=np.random.randn()
print("Randomly Initialized Slope:",m)
print("Randomly Initialized Intercept:",c)
```

```python
def predict(X,m,c):
  y_predict=m*X+c
  return y_predict
```

```python
y_pred=predict(X,m,c)
```

```python
print("First 5 predicted Values:")
print(y_pred[:5])
```

```python
def compute_cost(y,y_pred):
  m=len(y)
  cost=(1/(2*m))*np.sum((y_pred-y)**2)
  return cost
```

```python
cost=compute_cost(Y,y_pred)
print("Initial Cost(MSE):",cost)
```

```python
m=0.0
c=0.0

y_pred=predict(X,m,c)

cost=compute_cost(Y,y_pred)
print("Cost for m=",m,"and c=",c,"is:",cost)
```

```python
learning_rate=0.01
epochs=1000
print("Learning Rate:",learning_rate)
print("Epochs:",epochs)
```

```python
n=len(y)
```

```python
cost_history = [] # Initialize cost_history
for epoch in range(epochs):

  y_pred=predict(X,m,c)

  dm=(1/n)*np.sum(y_pred-Y)*X # Corrected division and used Y
  dc=(1/n)*np.sum(y_pred-Y) # Corrected division and used Y

  m=m-learning_rate*dm
  c=c-learning_rate*dc

  cost=compute_cost(Y,y_pred)
  cost_history.append(cost)
  if epoch%100==0:
    print(f"Epoch: {epoch}: Cost:{cost}")
```

```python
def predict_salary(years_experience,m,c,X_min,X_max):
  X=np.array(years_experience).reshape(-1,1)
  X_norm=(X-X_min)/(X_max-X_min)
  salary_pred=m*X_norm+c
  return salary_pred
```

```python
print(predict_salary(5,m,c,X_min,X_max))
```

```python
years=5
predicted_salary=predict_salary(years,m,c,X_min,X_max )
print(f"Predicted salary for {years} years of experience:")
print(predicted_salary[0][0])
```
